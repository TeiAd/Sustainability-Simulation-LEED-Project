```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define project phases and tasks with durations (in weeks)
phases = [
    ("Pre-Design & Planning", [
        ("Stakeholder Identification & Engagement", 2),
        ("Project Goals Definition", 1),
        ("LEED Credit Analysis & Target Setting", 1),
        ("Preliminary Budgeting & Feasibility Studies", 2),
        ("Site Selection & Initial Assessment", 2)
    ]),
    ("Design Development", [
        ("Architectural Design with Sustainability Integration", 4),
        ("Engineering Design for Building Systems", 3),
        ("LEED Documentation for Design Phase Credits", 2),
        ("Stakeholder Review & Approval of Design Plans", 1)
    ]),
    ("Construction Planning & Procurement", [
        ("Construction Timeline Development", 2),
        ("Contractor Selection & Onboarding", 3),
        ("Procurement of Sustainable Materials", 4),
        ("Construction Team Training on LEED Goals", 2)
    ]),
    ("Construction", [
        ("Site Preparation & Foundation Work", 4),
        ("Structural Construction", 6),
        ("Installation of Building Systems", 8),
        ("Sustainable Features Installation", 6),
        ("Ongoing LEED Compliance Checks", 36)  # Continuous
    ]),
    ("Commissioning & Testing", [
        ("Commissioning of Building Systems", 4),
        ("Indoor Environmental Quality Testing", 2),
        ("Final Inspections", 2)
    ]),
    ("Certification & Handover", [
        ("Submission of LEED Documentation", 2),
        ("LEED Certification Review Process", 4),
        ("Handover to Facility Management", 2)
    ])
]

# Set project start date
start_date = datetime(2024, 1, 1)

# Prepare data for the Gantt chart
tasks = []
start_dates = []
end_dates = []
colors = []
phase_colors = [
    "#FF9999", "#FFC000", "#FFFF99", "#99CC99", "#99CCFF", "#C299FF"
]  # Different colors for phases

for i, (phase_name, tasks_list) in enumerate(phases):
    phase_start = start_date if not end_dates else max(end_dates)
    for task_name, duration in tasks_list:
        task_start = phase_start if not tasks else max(end_dates)
        task_end = task_start + timedelta(weeks=duration)
        tasks.append(task_name)
        start_dates.append(task_start)
        end_dates.append(task_end)
        colors.append(phase_colors[i])

# Plot the Gantt chart
fig, ax = plt.subplots(figsize=(12, 8))
for i, task in enumerate(tasks):
    ax.barh(task, (end_dates[i] - start_dates[i]).days, left=start_dates[i], color=colors[i])

# Format the chart
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
ax.invert_yaxis()
ax.set_xlabel("Timeline")
ax.set_title("Gantt Chart for Sustainable Facility Construction Project")

plt.tight_layout()
plt.show()
```
